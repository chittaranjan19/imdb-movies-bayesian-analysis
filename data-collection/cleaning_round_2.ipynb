```python
def convert_time(duration):
    if 'hr' in duration:
        hours = duration.split()[0]
        if 'min' in duration:
            mins = duration.split()[2]
            return int(hours)*60 + int(mins)
    else:
        mins = duration.split()[0]
        return mins
    return int(hours) * 60    
    print(duration.index('min'))
```

```python
data = pd.read_csv('data2.csv')
data_copy = data
data_copy['release_date'] = data_copy['release_date'].str.replace('Release date', '')
data_copy['release_date'] = data_copy['release_date'].str.replace(r"\(.*?\)", "", regex=True)
data_copy['country'] = data_copy['country'].str.replace('Country of origin', '')
data_copy['country'] = data_copy['country'].str.replace('Countries of origin', '')
data_copy['budget'] = data_copy['budget'].str.replace('(estimated)', '')
data_copy['budget'] = data_copy['budget'].str.replace('(','')
data_copy['budget'] = data_copy['budget'].str.replace(')','')
data_copy.dropna(subset=['duration'], how='all', inplace=True)
data_copy.dropna(subset=['country'], how='all', inplace=True)
data_copy['duration'] = data_copy['duration'].apply(lambda x: convert_time(x))
data_copy['duration'] = pd.to_numeric(data_copy['duration'])
data_copy['main_country'] = data_copy['country'].apply(lambda x: get_main_country(x))
data_copy
```
